```python
import pandas as pd
```

```python
import plotly.express as px
```

```python
df_1 = pd.read_csv("data/wandb_export_2025-01-28T02_08_46.697+01_00.csv")
df_2 = pd.read_csv("data/wandb_export_2025-01-28T02_30_55.333+01_00.csv")
```

```python
df_1 = df_1.drop(columns=[col for col in df_1.columns if "__" in col])
df_2 = df_2.drop(columns=[col for col in df_2.columns if "__" in col])
```

```python
# make tables from wide to long
df_1 = df_1.melt(id_vars=["Step"], var_name="Algorithm", value_name="Global Step")
df_2 = df_2.melt(id_vars=["Step"], var_name="Algorithm", value_name="Success Rate")
```

```python
df_1 = df_1.dropna(subset=["Global Step"])
df_2 = df_2.dropna(subset=["Success Rate"])
```

```python
df_1.Algorithm = df_1.Algorithm.apply(lambda x: x.split("-")[0].strip())
df_2.Algorithm = df_2.Algorithm.apply(lambda x: x.split("-")[0].strip())
```

```python
df_2 = df_2.merge(df_1, on=["Step", "Algorithm"], how="inner")
```

```python
df_2.Algorithm = df_2.Algorithm.replace(
    {
        "eager": "CrossQ",
        "jumping": "SAC",
        "ruby": "TQC",
        "comfy": "DDPG",
        "solar": "TD3",
    }
)
```

```python
fig = px.line(
    df_2,
    x="Global Step",
    y="Success Rate",
    color="Algorithm",
    line_shape="spline",
    title="Success Rate over Global Step",
)
```

```python
fig.show()
fig.write_html("success_rate.html")
```
